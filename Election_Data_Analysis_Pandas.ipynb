```python
import pandas as pd
df = pd.read_csv("all_states_ge.csv")
df.head()
```

```python
print("Rows and Columns:")
print(df.shape)

print("\nColumns:")
print(df.columns)

df.info()
```

```python
df.isnull().sum()
```

```python
df_2019 = df[df["Year"] == 2019]

df_2019.head()
```

```python
state_votes = df_2019.groupby("State_Name")["Votes"].sum()

state_votes.sort_values(ascending=False)
```

```python
party_votes = df_2019.groupby("Party")["Votes"].sum()

party_votes.sort_values(
    ascending=False
    ).head(20)
```

```python
import matplotlib.pyplot as plt

state_votes.sort_values(ascending=False).head(10).plot(kind="bar")

plt.title("Top 10 States by Total Votes")
plt.show()
```

```python
party_votes = df_2019.groupby("Party")["Votes"].sum()

party_votes.sort_values(ascending=False).head(20)
```

```python
party_votes.sort_values(ascending=False).head(10).plot(kind="bar")

plt.title("Top 10 Parties by Votes")
plt.show()
```

```python
top_candidates = df_2019.sort_values(by="Votes",ascending=False)

top_candidates[
    ["Candidate",
         "Party",
              "State_Name",
                   "Votes"]
                   ].head(10)
```
